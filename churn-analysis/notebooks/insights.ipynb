```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
df = pd.read_csv("../data/Churn_Modelling.csv")
os.makedirs('../dashboard', exist_ok=True)

df.groupby('Geography')['Exited'].mean().plot(kind='bar')

plt.title('Taxa de Cancelamento por Pais')
plt.xlabel('Pais')
plt.ylabel('Taxa de Churn')

plt.savefig('../dashboard/churn_por_pais.png')
plt.show()

plt.close()
```

```python
df.groupby('Age')['Exited'].mean().plot()

plt.title('Taxa de cancelamento por idade ')
plt.xlabel('Idade')
plt.ylabel('Taxa de Churn')
plt.savefig('../dashboard/churn_por_pais.png')
plt.show()


plt.close()
```
